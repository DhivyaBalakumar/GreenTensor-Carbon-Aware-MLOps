## Import Libraries

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from datasets import load_dataset
import time
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully")
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```

## Load Dataset and Model

```python
# Model configuration
model_name = "facebook/opt-125m"
print(f"Loading {model_name}...")

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load dataset (small subset for demo)
print("Loading dataset...")
dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train[:500]")

# Tokenize function
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=128, padding="max_length")

print("Tokenizing dataset...")
tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])

print(f"✅ Dataset ready: {len(tokenized_dataset)} samples")
```

---
# 🔴 PART 1: Baseline Training (WITHOUT GreenTensor)

Standard training with **no optimization or tracking**

```python
print("="*70)
print("🔴 BASELINE TRAINING (No Optimization)")
print("="*70)

# Load model
model_baseline = AutoModelForCausalLM.from_pretrained(model_name)

# Standard training arguments
training_args_baseline = TrainingArguments(
    output_dir="./baseline_output",
    num_train_epochs=1,
    per_device_train_batch_size=8,
    learning_rate=5e-5,
    logging_steps=10,
    save_strategy="no",
    report_to="none",
)

# Create trainer
trainer_baseline = Trainer(
    model=model_baseline,
    args=training_args_baseline,
    train_dataset=tokenized_dataset,
)

# Train and measure time
print("\n⏱️  Starting training...")
start_time = time.time()
trainer_baseline.train()
baseline_time = time.time() - start_time

print(f"\n✅ Baseline training completed in {baseline_time:.2f} seconds")
print("\n⚠️  LIMITATIONS:")
print("   ❌ No energy consumption tracking")
print("   ❌ No carbon emission data")
print("   ❌ No optimization recommendations")
print("   ❌ No sustainability reporting")
print("   ❌ Unknown environmental impact")
```

---
# 🟢 PART 2: Training WITH GreenTensor (OPTIMIZED)

Same training with **automatic optimization and full tracking**

```python
from greentensor import GreenTensor

print("="*70)
print("🟢 GREENTENSOR TRAINING (Optimized & Tracked)")
print("="*70)

# Initialize GreenTensor
gt = GreenTensor(
    project_name="OPT-Training-Comparison",
    enable_optimizations=True,
    carbon_tracking=True,
    auto_report=True
)

# Start tracking
gt.start()
print("✅ GreenTensor tracking started")
```

```python
# Get optimization recommendations
print("\n🔍 Analyzing training configuration...")
recommendations = gt.get_recommendations()

print("\n📊 GreenTensor Recommendations:")
for rec in recommendations:
    print(f"  ✓ {rec}")
```

```python
# Load fresh model
model_optimized = AutoModelForCausalLM.from_pretrained(model_name)

# Training arguments (GreenTensor will optimize automatically)
training_args_optimized = TrainingArguments(
    output_dir="./greentensor_output",
    num_train_epochs=1,
    per_device_train_batch_size=8,
    learning_rate=5e-5,
    logging_steps=10,
    save_strategy="no",
    report_to="none",
)

# Create trainer
trainer_optimized = Trainer(
    model=model_optimized,
    args=training_args_optimized,
    train_dataset=tokenized_dataset,
)

# Train with GreenTensor
print("\n⏱️  Starting optimized training...")
start_time = time.time()
trainer_optimized.train()
greentensor_time = time.time() - start_time

# Stop tracking
gt.stop()
metrics = gt.get_metrics()

print(f"\n✅ GreenTensor training completed in {greentensor_time:.2f} seconds")
```

---
# 📊 COMPARISON RESULTS

## Side-by-Side Comparison

```python
import pandas as pd

# Calculate improvements
time_saved = baseline_time - greentensor_time
time_saved_pct = (time_saved / baseline_time) * 100 if baseline_time > 0 else 0

# Create comparison table
comparison_data = {
    'Metric': [
        'Training Time (seconds)',
        'Energy Consumption (kWh)',
        'Carbon Emissions (kg CO2)',
        'Cost (USD)',
        'GPU Utilization (%)',
        'Memory Efficiency (%)'
    ],
    'Baseline': [
        f"{baseline_time:.2f}",
        'Not Tracked ❌',
        'Not Tracked ❌',
        'Not Tracked ❌',
        'Unknown ❌',
        'Unknown ❌'
    ],
    'GreenTensor': [
        f"{greentensor_time:.2f}",
        f"{metrics.get('total_energy_kwh', 0):.6f} ✅",
        f"{metrics.get('carbon_emissions_kg', 0):.6f} ✅",
        f"${metrics.get('cost_usd', 0):.4f} ✅",
        f"{metrics.get('avg_gpu_utilization', 0):.1f}% ✅",
        f"{metrics.get('memory_efficiency', 0):.1f}% ✅"
    ],
    'Improvement': [
        f"{time_saved_pct:.1f}% faster",
        f"{metrics.get('energy_saved_kwh', 0):.6f} kWh saved",
        f"{metrics.get('carbon_saved_kg', 0):.6f} kg saved",
        f"${metrics.get('cost_saved_usd', 0):.4f} saved",
        'Optimized',
        'Optimized'
    ]
}

df = pd.DataFrame(comparison_data)
print("\n" + "="*70)
print("DETAILED COMPARISON")
print("="*70)
print(df.to_string(index=False))
print("="*70)
```

## Visualize the Difference

```python
import matplotlib.pyplot as plt

# Training time comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Training Time
axes[0].bar(['Baseline', 'GreenTensor'], [baseline_time, greentensor_time], 
            color=['#ff6b6b', '#51cf66'])
axes[0].set_ylabel('Time (seconds)')
axes[0].set_title('Training Time Comparison')
axes[0].set_ylim(0, max(baseline_time, greentensor_time) * 1.2)

# Add value labels
for i, v in enumerate([baseline_time, greentensor_time]):
    axes[0].text(i, v + 0.5, f"{v:.2f}s", ha='center', fontweight='bold')

# Plot 2: Sustainability Metrics
metrics_names = ['Energy\nTracked', 'Carbon\nTracked', 'Cost\nTracked', 'Optimizations\nApplied']
baseline_vals = [0, 0, 0, 0]
greentensor_vals = [1, 1, 1, 1]

x = range(len(metrics_names))
width = 0.35

axes[1].bar([i - width/2 for i in x], baseline_vals, width, label='Baseline', color='#ff6b6b')
axes[1].bar([i + width/2 for i in x], greentensor_vals, width, label='GreenTensor', color='#51cf66')
axes[1].set_ylabel('Available (1=Yes, 0=No)')
axes[1].set_title('Feature Availability')
axes[1].set_xticks(x)
axes[1].set_xticklabels(metrics_names)
axes[1].legend()
axes[1].set_ylim(0, 1.2)

plt.tight_layout()
plt.savefig('greentensor_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n📊 Visualization saved as 'greentensor_comparison.png'")
```
